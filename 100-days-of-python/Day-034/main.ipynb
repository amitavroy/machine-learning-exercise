# filtering by dates

```python
filt = (df['Date'] >  '2018') & (df['Date'] < '2019')
df.loc[filt].sort_values(by=['Date'], ascending=False)
```

```python
filt = (df['Date'] >=  '2020')
df.loc[filt].sort_values(by=['Date'], ascending=False)
```

```python
filt = (df['Date'] >=  pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt].sort_values(by=['Date'], ascending=False)
```
