```python
import pandas as pd
import numpy as np
import datetime as dt
```

Importing the data with custom field names so that it is easy 
to reference those fields later in the code

```python
desired_col_names = ["month_name", "month_str", "view", "views_inc", "watch_time_hrs", "watch_time_inc", "subs", "sub_inc", "revenue", "revenue_inc", "video_count"]

df = pd.read_csv('./../../data/youtube.csv', names=desired_col_names, header=0)
df.head()
```

# Data observation and clean up

Looking at the data and tweaking it based on requirements.
Right now, I don't plan to train any model. 

But, would like to analyse as much as possible and use some of the 
graph libraries to make certain observations

```python
df.describe()
```

```python
new_df = df.copy()

date_format = "%d/%m/%Y"

new_df["month_str"] = pd.to_datetime(new_df['month_str'], format=date_format)
new_df['view'] = new_df['view'].str.replace(",", "").astype(int)
new_df['watch_time_hrs'] = new_df['watch_time_hrs'].str.replace(",", "").astype(int)
new_df['watch_time_inc'] = new_df['watch_time_inc'].astype(int)
```
