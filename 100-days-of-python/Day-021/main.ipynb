# Analysing data

```python
df.head()
```

```python
country_count = df['Country'].value_counts(ascending = False)
country_count.head(10)
```

```python
countries_lt_1500 = country_count[country_count<1500]
countries_lt_1500
```

```python
def clean_country(value):
    if (value in countries_lt_1500):
        return 'Others'
    else:
        return value
```

```python
df['Country'] = df['Country'].apply(clean_country)
df['Country'].unique()
```

```python
df['Country'].value_counts(ascending=False)
```
